**Filtering & Aggregations (GROUP BY, HAVING)** :-

**4. Count the number of customers in each InternetService category.**

**5. Find the average MonthlyCharges for customers grouped by PaymentMethod.**

```python
query = """
SELECT paymentmethod, AVG(monthlycharges) AS AvgMonthlyCharges
FROM customer
GROUP BY paymentmethod;
"""
result = pd.read_sql(query, conn)
display(result)
```

**6. Show the top 5 tenure ranges (e.g., 0–12 months, 13–24 months, etc.) with the highest churn
count.**

```python
query = """
SELECT
    CASE
        WHEN tenure BETWEEN 0 AND 12 THEN '0-12 months'
        WHEN tenure BETWEEN 13 AND 24 THEN '13-24 months'
        WHEN tenure BETWEEN 25 AND 36 THEN '25-36 months'
        WHEN tenure BETWEEN 37 AND 48 THEN '37-48 months'
        WHEN tenure BETWEEN 49 AND 60 THEN '49-60 months'
        ELSE '61+ months'
    END AS TenureRange,
    COUNT(*) AS ChurnCount
FROM customer
WHERE churn = 'Yes'
GROUP BY TenureRange
ORDER BY ChurnCount DESC
LIMIT 5;
"""

result = pd.read_sql(query, conn)
display(result)
```

**7. Find payment methods where the average TotalCharges is greater than 3000.**

```python
query = """
SELECT paymentmethod, AVG(totalcharges) AS AvgTotalCharges
FROM customer
GROUP BY paymentmethod
HAVING avgtotalcharges > 3000;
"""

result = pd.read_sql(query, conn)
display(result)
```

**Unions** :-

**8. Write a query to list customers who are either on Month-to-Month contracts OR who have
churned (use UNION).**

```python
query = """
SELECT customerID, contract, churn
FROM customer
WHERE contract = 'Month-to-month'

UNION

SELECT customerID, contract, churn
FROM customer
WHERE churn = 'Yes';
"""

result = pd.read_sql(query, conn)
display(result)
```

**9. Use UNION to combine customer lists of those who have PhoneService and those who have
InternetService = Fiber optic.**

```python
query = """
SELECT customerID, phoneservice, internetservice
FROM customer
WHERE internetservice = 'Fiber optic'

UNION

SELECT customerID, phoneservice, internetservice
FROM customer
WHERE phoneservice = 'Yes';
"""

result = pd.read_sql(query, conn)
display(result)
```

**Views :-**

**14. Create a view that shows only customers who have churned with their tenure, monthly
charges, and contract type.**

```python
create_view = """
CREATE VIEW churnedcustomers AS
SELECT tenure, monthlycharges, contract
FROM customer
WHERE churn = 'Yes';

conn.execute(create_view)"""

query = "SELECT * FROM churnedcustomers;"
result = pd.read_sql(query, conn)
display(result)
```

**15. Create a view to calculate the churn rate for each contract type**

```python
create_view = """
CREATE VIEW churn_rate_by_contract AS
SELECT
    contract,
    COUNT(*) AS total_customers,
    SUM(CASE WHEN churn = 'Yes' THEN 1 ELSE 0 END) AS churned_customers,
    ROUND(
        (SUM(CASE WHEN churn = 'Yes' THEN 1 ELSE 0 END) * 100.0) / COUNT(*),
        2
    ) AS churn_rate_percentage
FROM customer
GROUP BY contract;
"""
query = "SELECT * FROM churn_rate_by_contract;"
result = pd.read_sql(query, conn)
display(result)
```

**16. Create a view that summarizes total revenue per InternetService.**

```python
create_view = """
CREATE VIEW revenue_per_internetservice AS
SELECT
    InternetService,
    SUM(TotalCharges) AS total_revenue
FROM customer
GROUP BY InternetService;
"""

conn.execute(create_view)
query = "SELECT * FROM revenue_per_internetservice;"
result = pd.read_sql(query, conn)
display(result)
```

**21. Calculate churn percentage grouped by gender.**

```python
query = """
SELECT
    gender,
    COUNT(*) AS total_customers,
    SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) AS churned_customers,
    ROUND(100.0 * SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) / COUNT(*), 2) AS churn_percentage
FROM Customer
GROUP BY gender;
"""
result = pd.read_sql(query, conn)
display(result)
```

**24. Compare the churn rate between customers who use Electronic check vs. other payment
methods.**

```python
query = """SELECT
    CASE
        WHEN PaymentMethod = 'Electronic check' THEN 'Electronic Check'
        ELSE 'Other Methods'
    END AS PaymentGroup,
    COUNT(*) AS total_customers,
    SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) AS churned_customers,
    ROUND(100.0 * SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) / COUNT(*), 2) AS churn_percentage
FROM customer
GROUP BY PaymentGroup;

"""
result = pd.read_sql(query, conn)
display(result)
```
